# <u>**Problem 1**</u>

### Plot the total profit across the months

* X label name = Month Number
* Y label name = Total profit



```python
# Your code here

# plottig the line chart to measure profit across months
plt.plot(company_sales['month_number'], company_sales['total_profit'], color='green', linestyle='solid', marker='o',
         label = 'monthly profit')

plt.title('Monthly Profit')
plt.xlabel('Month Number')
plt.ylabel("Total Profit")
plt.legend()

plt.show()
```

# <b><u>Problem 2 </u></b>

### Read face cream and facewash product sales data and show it using the bar chart. The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```python
# Your code here

# plotting bar chart to show units sold per month for facecream and facewash
plt.bar(np.arange(company_sales.shape[0]) - 0.4, company_sales['facecream'], color = 'red', label='Facecream', width=0.4)
plt.bar(np.arange(company_sales.shape[0]), company_sales['facewash'], color='green', label='Facewash', width=0.4)
plt.legend()
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.xticks(np.arange(company_sales.shape[0]), company_sales['month_number'])

plt.show()
```

# <b><u>Problem 3</u></b>

### Display the number of units sold per month for each product using multiline plots. (i.e. separate plotline for each product ).

```python
# Your code here

# plotting multiline plot to show units sold per month for each product
for column in company_sales.columns[1:7]:
    plt.plot(company_sales['month_number'], company_sales[column], label=column, marker='.', )
    plt.legend()
    plt.xlabel('Month Number')
    plt.ylabel('Units Sold')
```
