# <b>Problem 4 : You need to clean the data a little bit now. Replace all column values with the values <code>?</code> and <code>n.a</code> with <code>NaN</code>

```python
# Your code here

# replacing ? and na values with NaN
automobile_df = automobile_df.replace(['?', 'n.a'], np.nan)
```

# <b>Problem 5 : Print the company name which manufactured the most expensive car. Also print the price of the most expensive car

```python
# Your code here

# dropping null values from dataframe
cleaned_automobile_df = automobile_df.dropna(subset=['price'])

# changing price column datatype from object to int
cleaned_automobile_df['price'] = cleaned_automobile_df['price'].astype('int')
```

```python
# extracting most expensive car manufacturer and price of most expensive car
cleaned_automobile_df.groupby('make')['price'].max().sort_values(ascending=False).head(1)
```

# <b>Problem 6 : Calculate maximum horsepowers for each company. </b>

```python
# Your code here

# converting column horsepower from object to int
cleaned_automobile_df = cleaned_automobile_df.dropna(subset = ['horsepower'])
cleaned_automobile_df['horsepower'] = cleaned_automobile_df['horsepower'].astype('int')

# extracting maximum horsepower for each company
print('Maximum horsepower for each company')
cleaned_automobile_df.groupby('make')['horsepower'].max()
```

# <b> Problem 7 : Print the total number of cars manufactured by each company

```python
# Your code here

# extracting total cars made by each company
print('Number of Cars manufactured by each company')
cleaned_automobile_df['make'].value_counts()
```

# <b> Problem 8 : Create a new column for the updated price. This updated price will be calculated on top of the given price - if the engine is in front, price will be same else if the engine is in rear, price will be doubled </b>


```python
# Your code here. Use apply and lambda

cleaned_automobile_df['updated_price'] = cleaned_automobile_df.apply(lambda row : row['price']*2 if row['engine-location']  == 'rear' else row['price'], axis=1)

# checking whether the price for rear engine location is doubled or not
cleaned_automobile_df[cleaned_automobile_df['engine-location'] == 'rear']
```

# <b>Problem 11 : Concatenate the two data frames given below firstly row wise and secondly column wise.</b>

<code>GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}</code>

<code>japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}</code>

```python
# Your code here
GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}
japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}

german_cars_df = pd.DataFrame(GermanCars)

japanese_cars_df = pd.DataFrame(japaneseCars)

# concatenating two dataframes row wise
pd.concat((german_cars_df, japanese_cars_df), ignore_index=True, axis=0)

# concatenating two dataframes column wise
pd.concat((german_cars_df, japanese_cars_df), keys= ['German', 'Japanese'], axis=1)
```
